```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
salaries_data= pd.read_csv('/content/Salaries.csv')
salaries_data.head()
```

### **Part 1:Basic Data Exploration**

```python
#Display the number of rows and columns in the dataset
print("Number of rows: ",len(salaries_data), "\nNumber of columns: ", len(salaries_data.columns))

#Determine the data types of each column
print("\n\nChecking Data type: \n", salaries_data.dtypes)

#checking for missing values in each column
missing_values=salaries_data.isnull().sum()
print("\n\nChecking for missing values:\n", missing_values)
```

### **Part 2: Descriptive Statistics**

```python
#basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and finding the standard deviation
print("Mean Salary:", salaries_data['TotalPay'].mean() )
print("Median Salary:", salaries_data['TotalPay'].median() )
print("Mode Salary:", salaries_data['TotalPay'].mode() )

Smin=salaries_data['TotalPay'].min()
Smax= salaries_data['TotalPay'].max()
print("Minimum Salary:", Smin )
print("Maximum Salary:", Smax)
print("Range of Salaries:", Smax-Smin)

print("Standard Deviation of Salaries:",salaries_data['TotalPay'].std() )
```

### **Part 3: Data Cleaning**

```python
total_cells = np.product(salaries_data.shape)
total_missing = missing_values.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print("percent of data that is missing:  ", int(percent_missing),"%")

#17% of the cells in this dataset are empty
```
